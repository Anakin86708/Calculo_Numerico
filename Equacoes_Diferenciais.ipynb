```python
import numpy as np, scipy as sp, matplotlib.pyplot as mpl
from math import cos
```

# Equações Difirenciais

```python
m = 1
k = 1
x0 = 1
v0 = 0
tf = 10
n = 101

del_t = tf / (n-1)
```

```python
dif_v = lambda v,x:  v - (k/m)*x*del_t
dif_x = lambda v,x: x + v*del_t

vy = np.zeros(n)
xy = np.zeros(n)

vy[0] = v0
xy[0] = x0

# Utiliza as formulas de condição inicial para os cálculos
for i in range(0,n-1):
    vy[i+1] = dif_v(vy[i],xy[i])
    xy[i+1] = dif_x(vy[i],xy[i])
vy, xy
```

## Resultados obtidos

```python
x = np.linspace(0,tf,n)
fig, ax = mpl.subplots()
ax.plot(x,vy,label='V')
ax.plot(x,xy,label='X')
ax.legend()
#fig.show()
```

## Compara o resultado obtido com a função correta

```python
y = list(map(cos,x))
fig, ax = mpl.subplots()
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')
ax.spines['left'].set_smart_bounds(True)
ax.spines['bottom'].set_smart_bounds(True)
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.plot(x,xy,label='X')
ax.plot(x,y,label='Cos')
ax.legend()
#fig.show()
```
