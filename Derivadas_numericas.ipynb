```python
def derivada_numerica_var2(f,xi,xf,yi,yf,np):    
    pontos_x = num.linspace(xi,xf,np)
    pontos_y = num.linspace(yi,yf,np)
    
    resultado_derivada_central_x = []
    derivada_central_x = lambda j, i, x, y: (f(x[j+1],y[i]) - f(x[j-1],y[i])) / (x[j+1] - x[j-1])
    
    for indice_y, _ in enumerate(pontos_y):
        for indice_x, _ in enumerate(pontos_x):
            if indice_x != 0 and indice_x != len(pontos_x)-1:
                resultado_derivada_central_x.append(derivada_central_x(indice_x, indice_y, pontos_x, pontos_y))

    resultado_derivada_central_y = []
    derivada_central_y = lambda j, i, x, y: (f(x[j],y[i+1]) - f(x[j],y[i-1])) / (y[i+1] - y[i-1])
    
    for indice_x, _ in enumerate(pontos_x):
        for indice_y, _ in enumerate(pontos_y):
            if indice_y != 0 and indice_y != len(pontos_y)-1:
                resultado_derivada_central_y.append(derivada_central_y(indice_x,indice_y, pontos_x, pontos_y))

            
    return num.array(resultado_derivada_central_x), num.array(resultado_derivada_central_y)
```

```python
def ploting3D(x,y,f):
    x,y = num.meshgrid(x,y)

    fig = mpl.figure()
    ax = mpl.axes(projection="3d")

    ax.plot_wireframe(x,y,f(x,y), cmap='winter')
    mpl.show()
```
