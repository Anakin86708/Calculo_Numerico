```python
import matplotlib.pyplot as mpl
from math import pi, sin, cos, log
```

```python
def eixo_x(xi,xf):
    # Define todos os intevalos delta_x a partir da quantidade de pontos
    delta = lambda n: (xf-xi)/(n-1)
    delta_x = delta(np)
    

    # Calcula o número de pontos para cada delta_x e número de pontos
    pontos_x = []
    ponto_atual = lambda i: xi + delta_x * i

    for i in range(0,np):
        pontos_x.append(ponto_atual(i))
            
    return pontos_x
```

```python
def resultado(f,xi,xf,np):
    pontos_x = eixo_x(xi,xf)
    
    resultado_derivada_central = []
    derivada_central = lambda j, x: (f(x[j+1]) - f(x[j-1])) / (x[j+1] - x[j-1])
    
    for indice, _ in enumerate(pontos_x):
        if indice != 0 and indice != len(pontos_x)-1:
            resultado_derivada_central.append(derivada_central(indice, pontos_x))
            
    #print(resultado_derivada_central)
            # num.append(pontos_x[indice_np], atual_x(delta_x[indice_np])
    return resultado_derivada_central
```

```python
# TESTE
f = lambda x: sin(x)
np = 11
xi = 0
xf = 2*pi

y = resultado(f,xi,xf,np)
pontos_x = eixo_x(xi,xf)

y_sin = list(map(f,pontos_x))
mpl.plot(pontos_x,y_sin)
mpl.plot(pontos_x[1:-1],y)
mpl.show()
```
